```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = {
    "Category": ["A"] * 50 + ["B"] * 50 + ["C"] * 50,
    "Values": (
        list(np.random.normal(loc=5, scale=1, size=50)) +
        list(np.random.normal(loc=6, scale=1.5, size=50)) +
        list(np.random.normal(loc=7, scale=1, size=50))
    )
}
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.violinplot(x="Category", y="Values", data=df, palette="muted", scale="count")
plt.title("Multi-Data Series Violin Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
